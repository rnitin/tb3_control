Plot linear and angular error obtained from the waypoint navigation file

```python
import pickle
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open('list_er_lin.pkl', 'rb') as f:
    list_er_lin = pickle.load(f)
with open('list_er_ang.pkl', 'rb') as f:
    list_er_ang = pickle.load(f)
```

```python
# e_lin and e_ang simultaneously
plt.figure(figsize=(10,20))

plt.subplot(5,2,1)
plt.plot(np.array(list_er_lin[0]))
plt.title("path 1 - linear error")
plt.subplot(5,2,2)
plt.plot(np.array(list_er_ang[0]))
plt.title("path 1 - angular error")

plt.subplot(5,2,3)
plt.plot(np.array(list_er_lin[1]))
plt.title("path 2 - linear error")
plt.subplot(5,2,4)
plt.plot(np.array(list_er_ang[1]))
plt.title("path 2 - angular error")

plt.subplot(5,2,5)
plt.plot(np.array(list_er_lin[2]))
plt.title("path 3 - linear error")
plt.subplot(5,2,6)
plt.plot(np.array(list_er_ang[2]))
plt.title("path 3 - angular error")

plt.subplot(5,2,7)
plt.plot(np.array(list_er_lin[3]))
plt.title("path 4 - linear error")
plt.subplot(5,2,8)
plt.plot(np.array(list_er_ang[3]))
plt.title("path 4 - angular error")

plt.subplot(5,2,9)
plt.plot(np.array(list_er_lin[4]))
plt.title("path 5 - linear error")
plt.subplot(5,2,10)
plt.plot(np.array(list_er_ang[4]))
plt.title("path 5 - angular error")
```
